```python
def sample_io_to_text(sample_io):
    if not isinstance(sample_io, list):
        return ""
    texts = []
    for ex in sample_io:
        if isinstance(ex, dict):
            inp = ex.get("input", "")
            out = ex.get("output", "")
            texts.append(f"input: {inp} output: {out}")
    return " ".join(texts)
df["sample_io_text"] = df["sample_io"].apply(sample_io_to_text)
text_cols = [
    "title",
    "description",
    "input_description",
    "output_description"
]
df[text_cols] = df[text_cols].fillna("")
df["full_text"] = (
    df["title"] + " " +
    df["description"] + " " +
    df["input_description"] + " " +
    df["output_description"] + " " +
    df["sample_io_text"]
)
print(df["sample_io_text"].iloc[0])
print(df["full_text"].iloc[0][:400])
df.to_csv("../data/preprocessed_problems.csv", index=False)
```
