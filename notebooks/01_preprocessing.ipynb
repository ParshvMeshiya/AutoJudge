```python
import pandas as pd
from utils.preprocessing import clean_text

df = pd.read_csv("../data/problems.csv")

def sample_io_to_text(sample_io):
    if not isinstance(sample_io, list):
        return ""
    parts = []
    for ex in sample_io:
        if isinstance(ex, dict):
            parts.append(
                f"input: {ex.get('input','')} output: {ex.get('output','')}"
            )
    return " ".join(parts)

df["sample_io_text"] = df["sample_io"].apply(sample_io_to_text)

text_cols = [
    "title",
    "description",
    "input_description",
    "output_description"
]

df[text_cols] = df[text_cols].fillna("").applymap(clean_text)

df["full_text"] = (
    df["title"] + " " +
    df["description"] + " " +
    df["input_description"] + " " +
    df["output_description"] + " " +
    df["sample_io_text"]
)

df.to_csv("../data/preprocessed_problems.csv", index=False)
print("✅ Saved preprocessed_problems.csv")
```
